#### Merging Series and DataFrame to another DataFrame

```python
# Create a data frame with random values
df1 = pd.DataFrame(np.random.randn(5,4), index = ['A','B','C','D','E'], columns = list('WXYZ'))

# Create a new series with index (a.k.a., row) names same as that of above
arr = np.array([1,2,3,4,5])
df2 = pd.Series(arr, index = ['A','B','C','D','E']).rename('new')

# Merge the newly created series to the right of the dataframe, and also rename the new column name as 'new'
df1.merge(df2.to_frame(), left_index=True, right_index=True)
```

```python
# Create a data frame with random values
df1a = pd.DataFrame(np.random.randn(5,4), index = ['A','B','C','D','E'], columns = list('WXYZ'))

# Create another data frame with randome integres, and index (a.k.a., row) names same as that of above
df2a = pd.DataFrame(np.random.random_integers(1,5,size=(5,4)), index = ['A','B','C','D','E'], columns = ['NEW_W', 'NEW_X', 'NEW_Y', 'NEW_Z'])

# Merge the newly created series to the right of the dataframe, and also rename the new column name as 'new'
df1a.merge(df2a, left_index=True, right_index=True)
```
